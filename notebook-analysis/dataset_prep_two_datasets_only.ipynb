```python
import pandas as pd

# Load the two datasets
patient_df = pd.read_csv("../data/data_clinical_patient.txt", sep="\t", comment="#")
sample_df  = pd.read_csv("../data/data_clinical_sample.txt", sep="\t", comment="#")

# Merge them using PATIENT_ID
full_df = patient_df.merge(sample_df, on="PATIENT_ID", how="inner")

# Create the target column
full_df["recurred"] = full_df["RFS_STATUS"].map({"1:Recurred": 1, "0:Not Recurred": 0})

# Show basic information
print(f"Shape: {full_df.shape}")
print(f"\nTarget:\n{full_df['recurred'].value_counts(dropna=False)}")
print(f"\nUnique patients: {full_df['PATIENT_ID'].nunique()}")
print(f"One row per patient: {full_df['PATIENT_ID'].nunique() == len(full_df)}")
print(f"\nMissing values:\n{full_df.isnull().sum()[full_df.isnull().sum() > 0].sort_values(ascending=False)}")

full_df.head()
```
